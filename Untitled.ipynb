### First we need to get the relation between columns which contains missing values and the target column 

```python
dataset.columns
```

```python
corr = dataset['education'].corr(dataset['TenYearCHD'])
print("The correlation between education column and the target column = ", corr)
```

```python
corr = dataset['cigsPerDay'].corr(dataset['TenYearCHD'])
print("The correlation between cigerattes per day column and the target column = ", corr)
```

```python
corr = dataset['BPMeds'].corr(dataset['TenYearCHD'])
print("The correlation between blood pressure column and the target column = ", corr)
```

```python
corr = dataset['totChol'].corr(dataset['TenYearCHD'])
print("The correlation between total cholesterol level column and the target column = ", corr)
```

```python
corr = dataset['BMI'].corr(dataset['TenYearCHD'])
print("The correlation between Body Mass Index column and the target column = ", corr)
```

```python
corr = dataset['heartRate'].corr(dataset['TenYearCHD'])
print("The correlation between heart rate column and the target column = ", corr)
```

```python
corr = dataset['glucose'].corr(dataset['TenYearCHD'])
print("The correlation between glucose column and the target column = ", corr)
```

### According to the correlation between missing values column and the target value , it seems that all of those columns have positive correlation with the target except the education column which has negative correlation, so all of these columns are important in the model accuracy
